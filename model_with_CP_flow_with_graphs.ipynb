```python
# Load the CP_results
with open(f"{args.save_path}/CP_results.pkl", "rb") as f:  # <- update this path
    cp_results = pickle.load(f)

# Extract data
coords = testset.coords
x = coords[:, 0]
y = coords[:, 1]
pred = np.squeeze(test_preds_norm)
lower = np.squeeze(np.array([item['lower'] for item in cp_results]))
upper = np.squeeze(np.array([item['upper'] for item in cp_results]))
print(lower.shape)

# Create the 3D plot
fig = plt.figure(figsize=(14, 9))
ax = fig.add_subplot(111, projection='3d')

# Prediction surface
sc = ax.scatter(x, y, pred, c=pred, cmap='viridis', label='Prediction', s=25)

# Confidence interval surfaces
ax.plot_trisurf(x, y, lower, color='red', alpha=0.3, label='Lower Bound')
ax.plot_trisurf(x, y, upper, color='green', alpha=0.3, label='Upper Bound')

# Axis labels and legend
ax.set_xlabel("X Coordinate")
ax.set_ylabel("Y Coordinate")
ax.set_zlabel("Predicted Value")
ax.set_title("3D Interpolation with Conformal Prediction Intervals")

# Add colorbar and legend
fig.colorbar(sc, ax=ax, shrink=0.5, aspect=10, label='Predicted Value')
plt.legend()
plt.show()
```
